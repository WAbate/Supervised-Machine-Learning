```python
# Import Dependancies
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Read in CSV
train_df = pd.read_csv(Path('Instructions/Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Instructions/Resources/2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data
train_df = (train_df
                .replace({'loan_status':{'low_risk':'1'}})
                .replace({'loan_status':{'high_risk':'0'}})
                .replace('w',0)
                .replace('f',1)
                .replace('Individual','1')
                .replace('Joint App','0')
                .replace('ANY','0')
                .replace('RENT','1')
                .replace('MORTGAGE','2')
                .replace('OWN','3')
                .replace('Y','1')
                .replace('N','0')
                .replace('Y','1')
                .replace('N','0')
                .replace('Not Verified','0')
                .replace('Source Verified','1')
                .replace('Verified','1'))
```

```python
# Drop columns and show with .head()
train_df.drop(['index','pymnt_plan'], axis='columns', inplace=True)
train_df.drop(['Unnamed: 0'], axis='columns', inplace=True)
train_df.dropna()
train_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
test_df =(test_df
             .replace({'loan_status':{'low_risk':'1'}})
             .replace({'loan_status':{'high_risk':'0'}})
             .replace('w',0)
             .replace('f',1)
             .replace('Individual','1')
             .replace('Joint App','0')
             .replace('ANY','0')
             .replace('RENT','1')
             .replace('MORTGAGE','2')
             .replace('OWN','3')
             .replace('Y','1')
             .replace('N','0')
             .replace('Y','1')
             .replace('N','0')
             .replace('Not Verified','0')
             .replace('Source Verified','1')
             .replace('Verified','1'))
```

```python
# Drop columns and show with .head()
test_df.drop(['index','pymnt_plan'],axis='columns', inplace=True)
test_df.drop(['Unnamed: 0'],axis='columns', inplace=True)
test_df.dropna()
test_df.head()
```

```python
# add missing dummy variables to testing set
# Train the Logistic Regression model on the unscaled data and print the model score


X_train = train_df.drop('loan_status', axis = 'columns')
y_train = train_df['loan_status'].values
X_test = test_df.drop('loan_status', axis='columns')
y_test = test_df['loan_status'].values

classifier = LogisticRegression(solver='lbfgs',max_iter=200000)
classifier.fit(X_train, y_train)
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")

y_pred = classifier.predict(X_test)
print(f'Logistic Regression Model Accuracy: {100 * accuracy_score(y_test, y_pred):.2f}%')
```

```python
# Train a Random Forest Classifier model and print the model score

rfc = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)

print(f'Training Score: {rfc.score(X_train, y_train)}')
print(f'Testing Score: {rfc.score(X_test, y_test)}')
```

```python
# Scale the data

s_scaler = StandardScaler()
X_train_scale = s_scaler.fit_transform(X_train)
X_test_scale = s_scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier.fit(X_train_scale, y_train)

print(f"Training Data: {classifier.score(X_train_scale, y_train)}")
print(f"Testing Data: {classifier.score(X_test_scale, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
print(f'Training Score: {rfc.score(X_test_scale, y_test)}')
print(f'Testing Score: {rfc.score(X_train_scale, y_train)}')
```
